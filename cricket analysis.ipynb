# T20 CRICKET WORLD CUP DATA ANALYSIS

```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

# Extracting web page using BeautifulSoup

```python
soup=requests.get('https://stats.espncricinfo.com/ci/engine/records/team/match_results.html?id=14450;type=tournament').text
```

```python
page=BeautifulSoup(soup,'lxml')
```

```python
#you can use "print(page.prettify())" to see the entire webpage that has been scrapped 
```

# web scraping

```python
li=page.find_all('tr')
```

# Match Summary Table

```python
ui=[]
links_stats=[]
for i in range(1,len(li)):
    temp=li[i].find_all('td')
    temp1={
        'Team1': temp[0].text.strip(),
        'Team2': temp[1].text.strip(),
        'Winner': temp[2].text.strip(),
        'margin': temp[3].text.strip(),
        'Ground': temp[4].text.strip(),
        'Match Date': temp[5].text.strip(),
        'Score Id':temp[6].text.strip(),
    }
    links_stats.append('https://www.espncricinfo.com'+temp[6].find('a').get('href'))
    ui.append(temp1)

df_matchSummary = pd.DataFrame.from_records(ui)
```

```python
df_matchSummary
```

# Batsman Summary

```python
bat=[]
extra=[]
bat_stats=[]

for i in links_stats:
    incr=0
    soup1=requests.get(i).text
    pg = BeautifulSoup(soup1, 'lxml')
    
    y = pg.find_all('table',class_='ci-scorecard-table')
    z=pg.find_all('span',class_='ds-capitalize') # batting table
    for j in y:
        extra.append([j.find_next_sibling(),z[0].text.strip() + ' vs ' + z[1].text.strip(),z[not incr].text.strip()])
        # bowling table
        tr_bat1=j.find_all('tr',class_='')   #finding all rows of a table
        for k in range(1,len(tr_bat1)):
            temp_td=tr_bat1[k].find_all('td')
            if len(temp_td)==8:
                temp2={
                    'Match':z[0].text.strip() + ' vs ' + z[1].text.strip(),
                    'Batsmen':temp_td[0].text.strip(),
                    'Dismissal':temp_td[1].text.strip(),
                    'Batting pos':k,
                    'Team Innings':z[incr].text.strip(),
                    'R':temp_td[2].text.strip(),
                    'B':temp_td[3].text.strip(),
                    '4s':temp_td[5].text.strip(),
                    '6s':temp_td[6].text.strip(),
                    'SR':temp_td[7].text.strip(),
                }
                bat_stats.append('https://www.espncricinfo.com'+temp_td[0].find('a').get('href'))
                
                bat.append(temp2)
        incr+=1
```

```python
df_batting = pd.DataFrame.from_records(bat)
df_batting.head()
```

# Bowling Summary

```python
bowl=[]
bowl_stats=[]
for i in extra:
    bo=i[0].find_all('tr',class_='')
    for j in range(1,len(bo)):
        temp_td=bo[j].find_all('td')
        temp={
            'Match':i[1],
            'Bowling Team':i[2],
            'Bowler':temp_td[0].text.strip(),
            'O':temp_td[1].text.strip(),
            'M':temp_td[2].text.strip(),
            'R':temp_td[3].text.strip(),
            'W':temp_td[4].text.strip(),
            'ECO':temp_td[5].text.strip(),
            'WD':temp_td[9].text.strip(),
            'NB':temp_td[10].text.strip()
        }
        bowl_stats.append('https://www.espncricinfo.com'+temp_td[0].find('a').get('href'))
        bowl.append(temp)
```

```python
df_bowling=pd.DataFrame.from_records(bowl)
df_bowling.head()
```

# Players Information

```python
Name = []
Age = []
Batting_Style = []
Bowling_Style = []
Playing_Role = []
count = 1

for i in bat_stats:
    soup2=requests.get(i).text
    pg1 = BeautifulSoup(soup2, 'lxml')
    try:
        ho=pg1.find('div', class_ = 'ds-p-4')
        child = ho.find('div').find('div').find_all('div')
        for j in child:
            info = j.find_all('p')
            if(info[0].text.strip()=='Full Name'):
                try:
                    Name.append(info[1].text.strip())
                except:
                    Name.append(np.nan)
            elif(info[0].text.strip()=='Age'):
                try:
                    Age.append(info[1].text.strip())
                except:
                    Age.append(np.nan)
            elif(info[0].text.strip()=='Batting Style'):
                try:
                    Batting_Style.append(info[1].text.strip())
                except:
                    Batting_Style.append(np.nan)
            elif(info[0].text.strip()=='Bowling Style'):
                try:
                    Bowling_Style.append(info[1].text.strip())
                except:
                    Bowling_Style.append(np.nan)
            elif(info[0].text.strip()=='Playing Role'):
                try:
                    Playing_Role.append(info[1].text.strip())
                except:
                    Playing_Role.append(np.nan)
        if (len(Batting_Style) < count):
                Batting_Style.append('N/A')
        if (len(Bowling_Style) < count):
            Bowling_Style.append('N/A')
        if (len(Playing_Role) < count):
            Playing_Role.append('N/A')
        count = count + 1
    except:
        continue
    
dict = {'Name':Name, 'Age':Age, 'Batting_Style': Batting_Style, 'Bowling_Style':Bowling_Style, 'Playing_Role':Playing_Role}
```

```python
df_player_info_1=pd.DataFrame(dict)
df_player_info_1.head()
```

```python
df_player_info_2=pd.DataFrame(dict2)
df_player_info_2.head()
```

# Data Cleaning

```python
df_batting['out/notout']=df_batting['Dismissal'].apply(lambda x: 'not out' if x == 'not out' else 'out')
```

```python
df_batting.head()
```

```python
#applying regex to find special charachters
find=[]
for row in df_batting['Batsmen']:
    find.append(re.findall('[^A-Za-z ]',row))
```

```python
#finding batsmen containing \xa0
for i in range(0,len(find)):
    if len(find[i])>0 and find[i][0]=='\xa0':
        print(df_batting['Batsmen'][i])
        
```

```python
#removing special characters from names
df_batting['Batsmen']=df_batting.Batsmen.apply(lambda x:x.replace('†',''))
df_batting['Batsmen']=df_batting.Batsmen.apply(lambda x:x.replace('\xa0',''))
df_batting['Batsmen']=df_batting.Batsmen.apply(lambda x:x.replace('(c)',''))
```

```python
df_batting['Match'] = df_batting['Match'].apply(lambda x: x.replace('United Arab Emirates','U.A.E.'))
df_batting['Team Innings'] = df_batting['Team Innings'].apply(lambda x: x.replace('United Arab Emirates','U.A.E.'))
```

```python
df_batting.drop(columns=['Dismissal'],inplace=True)
```

```python
df_batting.head()
```

```python
match_dict={}

for index,row in df_matchSummary.iterrows():
    key1=row['Team1'] + ' vs ' + row['Team2']
    key2=row['Team2'] + ' vs ' + row['Team1']
    match_dict[key1] = row['Score Id']
    match_dict[key2] = row['Score Id']
```

```python
df_batting['id']=0
for index,row in df_batting.iterrows():
    #print(1)
    for key in match_dict:
        if row['Match']==key:
            df_batting['id'][index]=match_dict[key]
df_batting['id'].head(20)
```

```python
batsman = df_batting.drop(columns=['Match','Batting pos','Team Innings','SR','out/notout','id'],axis=1)
```

```python
#changing data type of required columns from string to integer 
batsman['R'] = batsman['R'].astype(int)
batsman['B'] = batsman['B'].astype(int)
batsman['4s'] = batsman['4s'].astype(int)
batsman['6s'] = batsman['6s'].astype(int)
batsman.dtypes
```

```python
#grouping records according to Batsmen column
batsman_stats = batsman.groupby('Batsmen').sum()
```

```python
batsman_stats['SR'] = batsman_stats['R']/batsman_stats['B']*100
```

```python
batsman_stats = batsman_stats.sort_values(by='R', ascending=False).reset_index()
```

```python
df_bowling['Bowler'].nunique()
```

```python
bowler = df_bowling.drop(columns=['Match','Bowling Team','ECO'], axis=1)
```

```python
bowler.dtypes
```

```python
bowler['O'] = bowler['O'].astype(float)
bowler['M'] = bowler['M'].astype(float)
bowler['R'] = bowler['R'].astype(int)
bowler['W'] = bowler['W'].astype(int)
bowler['WD'] = bowler['WD'].astype(int)
bowler['NB'] = bowler['NB'].astype(int)
bowler.dtypes
```

```python
bowler['B'] = bowler['O'].apply(np.floor)*6 + (bowler['O']-bowler['O'].apply(np.floor))*10
bowler['B'] = bowler['B'].astype(int)
bowler.drop(columns=['O'],axis=1,inplace=True)
bowler.head()
```

```python
bowler.drop(columns=['M'],axis=1,inplace=True)
```

```python
bowler_stats = bowler.groupby('Bowler').sum()
bowler_stats['O'] = bowler_stats['B'].floordiv(6) + (bowler_stats['B']-bowler_stats['B'].floordiv(6)*6)*0.1
bowler_stats['ECO'] = (bowler_stats['R']/bowler_stats['O']).round(2)
bowler_stats.head()
```

```python
bowler_stats = bowler_stats.sort_values(['W','ECO'],ascending=[False,True]).reset_index()
```
